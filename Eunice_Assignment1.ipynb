```python
#How many spend categories are there?
df_Spend.groupby("Type").sum()
df_Spend.dropna(axis=1, )
```

```python
#How many spend categories are there?
df_Spend.groupby("Type").size()
#df_Spend["Type"].value_counts()
```

```python
 #Which category has the highest average spend?
#df_Spend["Type"].value_counts()
category_Highest_Average=df_Spend.groupby("Type").agg (averageSpend=("Amount",np.mean))
category_Highest_Average.sort_values("averageSpend", ascending=False)
```

```python
#What is the average monthly spent by product categories?
#df_Spend.dtypes
#this is just the averahe monthly spend
df_Spend.groupby(df_Spend.Month.dt.month)["Amount"].mean() 
```

```python
# Which customers are reaching 90% or more of their spending limit?
df_spend_Limit=pd.merge(df_Customer_Details,df_Spend,left_on="Customer",right_on="Customer",how='inner')  
df_spend_Limit

```

```python
#Which city has the maximum number of spenders, each month? Is there a need to run campaigns in specific cities?
spenderCity=df_spend_Limit.groupby(["City"], as_index=False)["Amount"].max() 
spenderCity.sort_values
```

```python
#Which age group spends the most?
groupSpend=df_spend_Limit.groupby("Age").agg (averageSpend=("Amount",np.mean))
groupSpend.sort_values("averageSpend", ascending=False)
```
